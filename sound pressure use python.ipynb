```python
audio_file = '/Users/jinxuanwu/tmp/audio.csv'
```

```python
with open(audio_file) as f:
    measures = list(map(float, f.readlines()))
```

```python
measures
```

```python
import math, statistics
print ("raw sound pressure is: " + str(20 * math.log10(statistics.stdev(measures)) - 20*math.log10(20e-6)))
```

```python
#Apply A weighting Filter
import numpy as np
measure_len = len(measures)
num_uniq_pts = math.ceil((measure_len + 1) / 2)
measures_fft = np.fft.fft(measures)
measures_fft_first_half = measures_fft[:num_uniq_pts]
sample_rate = 22050

f = (np.arange(num_uniq_pts)) * sample_rate / measure_len

c1 = 3.5041384e16
c2 = pow(20.598997, 2)
c3 = pow(107.65265, 2)
c4 = pow(737.86223, 2)
c5 = pow(12194.217,2)

f = f ** 2
num = c1 * np.power(f, 4)
den = ((c2 + f)**2) * (c3 + f) * (c4 + f) * ((c5+f)**2)
A = num / den

XA = measures_fft_first_half * A

print (len(XA))
if measure_len % 2 == 1:                 
    XA = np.concatenate((XA, np.conjugate(XA[-1:1:-1])), axis = 0)
else:
    #print (len(XA), len(np.conjugate(XA[-1:0:-1])))
    XA = np.concatenate((XA, np.conjugate(XA[-1:1:-1])), axis = 0);

XA = np.real(np.fft.ifft(XA))

xA = np.reshape(XA, measure_len)

print ("sound pressure is: " + str(20 * math.log10(statistics.stdev(xA)) - 20*math.log10(20e-6)))
```
